## Load data

```python
df = pd.read_csv('../data/raw/gt.txt', sep=';', header=None)

df.head()
```

## GTSDB to YOLO format

```python
images_dir = "../data/raw/"

df_yolo = gtsdb_to_yolo.gtsdb_to_yolo(df, images_dir)
```

## Train Test Split

```python
grouped = df_yolo.groupby("filename")
image_files = list(grouped.groups.keys())

random.shuffle(image_files)
```

```python
len(image_files)
```

```python
train_files = image_files[:500]
val_files   = image_files[500:650]
test_files  = image_files[650:900]
```

```python
images_src = "../data/raw"
base_out = "../data/processed"

tts.process_split(train_files, "train", grouped, images_src, base_out)
tts.process_split(val_files, "val", grouped, images_src, base_out)
tts.process_split(test_files, "test", grouped, images_src, base_out)
```
